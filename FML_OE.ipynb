## Prediction of Price of an Estate by it's Size using LINEAR REGRESSION.

Importing Libraries

```python
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

Reading data file and making a dataframe of it

```python
data = pd.read_csv('Estate_Prices.csv')
df = pd.DataFrame(data)
```

```python
df = pd.DataFrame(data,columns=['Area','Price'])
print(df.head(10))
```

Scatter Plot of Training and Testing Data Set

```python
plt.xlabel('Area of Estate')
plt.ylabel('Price of Estate')
plt.title('DataSet')
x = df[['Area']]
y = df['Price']
reg = linear_model.LinearRegression()
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
reg.fit(x_train,y_train)
plt.scatter(x_train,y_train,label="Training set",color='r')
plt.scatter(x_test,y_test,label="Testing set",color='g')
plt.legend()
plt.show()
```

Scatter Plot of Training Data Set 

```python
y_pred = reg.predict(x_test)
plt.plot(x_test,y_pred,color="dodgerblue",label="Linear Regression ")
plt.scatter(x_train,y_train,c="r",label="Data Points")
plt.title("Training Data Set")
plt.xlabel("Area of Estate")
plt.ylabel("Price of Estate")
plt.legend()
plt.show()
```

Scatter Plot of Testing Data Set

```python
result = reg.predict(x_test)
y_pred = reg.predict(x_test)
plt.plot(x_test,result,label='Linear regression',color='dodgerblue')
plt.scatter(x_test,y_test,label="test set",color='g')
plt.title("Testing Data Set")
plt.xlabel("Area of Estate")
plt.ylabel("Price")
plt.legend()
plt.show()
```

Scatter Plot of Correct and Predicted Values

```python
plt.scatter(x_test,y_test,c="tomato",label="Correct Values")
plt.scatter(x_test,y_pred,c="lawngreen",label="Predicted Values")
plt.title("Difference plot")
plt.xlabel("Area of Estate")
plt.ylabel("Price")
plt.legend()
plt.show()
```

Taking user input for predicting the Dependent Variable (Price) and showing other parameters of the model.

```python
x = float(input("Enter the area of Estate: "))
predicted_price = reg.predict([[x]])
print(f'The predicted Price for an Estate of size {x} yards is: ',*predicted_price)
print('Intercept:',reg.intercept_)
print('Slope: ',*reg.coef_)
accuracy=reg.score(x_test,y_test)
print('Accuracy of the model is: ',accuracy)
```
